```python
import matplotlib.pyplot as plt
from PIL import Image
import os

# Load all image file paths from the 'generated_blocks' folder
generated_blocks_dir = "generated_blocks"
generated_blocks = [
    os.path.join(generated_blocks_dir, fname)
    for fname in os.listdir(generated_blocks_dir)
    if fname.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))
]

# Resize images to 512x512
resized_images = [Image.open(img).resize((512, 512), resample=Image.NEAREST) if isinstance(img, str) else img.resize((512, 512), resample=Image.NEAREST) for img in generated_blocks]

# Plot images in 2 columns
n_images = len(resized_images)
n_cols = 2
n_rows = (n_images + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 5 * n_rows))
axes = axes.flatten()

for ax, img in zip(axes, resized_images):
    ax.imshow(img)
    ax.axis('off')

# Hide any unused subplots
for ax in axes[len(resized_images):]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
